```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# importar el seprador de muestras para entrenamiento y pruebas 

from sklearn.model_selection import train_test_split
```

```python
# conjunto de datos disponibless con seaborn 
sns.get_dataset_names()
```

```python
# cargar el dataset
data = sns.load_dataset('iris')
data.head(5)
```

```python
data.describe()
```

```python
# DISTRIBUCION DE CADA CLASE
data.groupby('species').size()
```

```python
# Separacion de datos en un conjunto para neetrenamiento y otro para pruebas

train, test = train_test_split(data,test_size=0.4, stratify=data['species'],random_state=8)
```

# An√°lisis exploratorio de datos

```python
train.describe()
```

```python
train.groupby('species').size()
```

```python
# Histogramas para observar la distribucion de los 4 atributos 
n_bins = 10
fig, axs = plt.subplots(2, 2)
axs[0,0].hist(train['sepal_length'], bins=n_bins)
axs[0,0].set_title('Longitud  de sepalo')
axs[0,1].hist(train['sepal_width'], bins=n_bins)
axs[0,1].set_title('Ancho  de sepalo')
axs[1,0].hist(train['petal_length'], bins=n_bins)
axs[1,0].set_title('Longitud  de petalo')
axs[1,1].hist(train['petal_width'], bins=n_bins)
axs[1,1].set_title('Ancho de petalo ')
plt.tight_layout(pad=1.0);
plt.show()
```

```python
# Diagrama Box-Plot

figl, axsl = plt.subplots(2,2)
fn = ['sepal_lenght','sepal_width','petal_lenght', 'petal_width']
cn = ['setosa', 'versicolor', 'virginica']
sns.boxplot (x='species', y='sepal_length', data=train, order=cn, ax=axsl[0,0])
sns.boxplot (x='species', y='sepal_width', data=train, order=cn, ax=axsl[0,1])
sns.boxplot (x='species', y='petal_length', data=train, order=cn, ax=axsl[1,0])
sns.boxplot (x='species', y='petal_width', data=train, order=cn, ax=axsl[1,1])
figl.tight_layout(pad=1.0);
```

```python
#Diagrama de violin
sns.violinplot(x='species', y='petal_length',hue='species',
                data=train,order=cn, palette='colorblind');
```

```python
#Matriz de correlacion 
cormat =  train.corr(numeric_only=True)
```

```python
cormat
```

```python
#Mapa de calor
sns.heatmap(cormat, annot=True, square=True)
```

```python
#Diagrama de dispersion de los atributos emparejadaos 
sns.pairplot(train, hue='species', height=2, palette='colorblind')
```
