## Clasificador Naive Bayes


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics
```

```python
# Cargar el dataset
data = sns.load_dataset('iris')
train,test = train_test_split(data,test_size=0.4,stratify=data['species'],
                                random_state=29)
```

```python
# Separacion de la etiqueta de clase y atributos

X_train = train[['sepal_length','sepal_width',
                 'petal_length','petal_width']]
y_train = train['species']
X_test = test[['sepal_length','sepal_width',
               'petal_length','petal_width']]
y_test = test['species']
```

```python
# Clasificador Gaussiano Naive Bayes

mod_gnb = GaussianNB()
y_pred = mod_gnb.fit(X_train,y_train).predict(X_test)
print('La prediccion del clasificador en los datos de prueba es {:0.3f}'
      .format(metrics.accuracy_score(y_pred, y_test)))
```

```python
# Clasificador GNB utilizando solamente 2 características
columnas = ['petal_length', 'petal_width']
mod_gnb2 = GaussianNB()
y_pred2 = mod_gnb2.fit(X_train[columnas], y_train).predict(X_test[columnas])
print('La predicción del clasificador en los datos de prueba es {:.3f}'
      .format(metrics.accuracy_score(y_pred2, y_test)))
```
