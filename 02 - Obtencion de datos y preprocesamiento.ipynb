# WebScraping con BeautifulSoup

```python
import requests
import pandas as pd
from io import StringIO
from bs4 import BeautifulSoup
```

```python
# Enlace a ala pagina
url = 'https://es.wikipedia.org/wiki/Organizaci%C3%B3n_territorial_de_Guatemala'
```

```python
response = requests.get(url)
```

```python
print('Status code ', url, ':', response.status_code)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
tabla = soup.find('table', {'class':'wikitable'})
```

```python
tabla
```

```python
tabla = pd.read_html(StringIO(str(tabla)))
tabla
```

```python
departamento = tabla[1]
departamento
```

```python
departamento.columns = ['Departamento','Cabecera','Superficie','Poblacion']
departamento
```

```python
departamento.info()
```

```python
departamento.describe()
```

```python
a = departamento.Superficie[0]
a
```

```python
a = a.replace('\xa0','')
int(a)
```

```python
departamento['Superficie']= list(int(x.replace('\xa0','')) for x in departamento['Superficie'])
```

```python
departamento['Poblacion']= list(int(x.replace('\xa0','')) for x in departamento['Poblacion'])
```

```python
departamento
```
